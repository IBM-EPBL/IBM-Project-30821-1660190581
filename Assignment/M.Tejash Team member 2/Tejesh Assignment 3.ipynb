# 6. Find the outliers and replace the outliers

```python
sns.boxplot(data['Diameter'])
```

```python
data.mean()
```

```python
data.std()
```

```python
q=data.quantile(q=[0.75,0.5])
iqr=q.iloc[0]-q.iloc[1]
iqr
```

```python
l=q.iloc[1]-(1.5*iqr)
l['Diameter']
```

```python
u=q.iloc[1]+(1.5*iqr)
u['Diameter']
```

```python
sns.boxplot(data['Diameter'])
```

```python
data['Diameter']=np.where(data['Diameter']>u['Diameter'],u['Diameter'],np.where(data['Diameter']<l['Diameter'],l['Diameter'],data['Diameter']))
sns.boxplot(data['Diameter'])
```
