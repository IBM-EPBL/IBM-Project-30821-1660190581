# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
file=pd.read_csv("C:/Users/MURALI/Downloads/Admission_Predict.csv")
data=pd.DataFrame(file)
```

```python
plt.scatter(data.index,data['GRE Score'])
plt.show()
```

```python
sns.histplot(data["GRE Score"],color='blue')
```

```python
sns.displot(data["GRE Score"],color='yellow')
```

```python
sns.boxplot(data["GRE Score"],color='red')
```

```python
plt.scatter(data.SOP, data.CGPA)
plt.title('SOP  vs. CGPA')
plt.xlabel(' SOP')
plt.ylabel('CGPA')
```

```python
sns.barplot(data["SOP"],data["CGPA"])
```

```python
sns.lineplot(data["SOP"],data["CGPA"])
```

```python
sns.scatterplot(x=data.SOP,y=data.CGPA,color='green')
```

```python
sns.regplot(data['SOP'],data['CGPA'],color='pink')
```

```python
data.corr()
```

```python
pd.plotting.scatter_matrix(data.loc[:,"GRE Score":"Research"],diagonal="kde",figsize=(20,15))
plt.show
```
